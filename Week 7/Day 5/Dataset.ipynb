```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv ('supermarket_sales.csv')
df.head() # Display the first 5 rows of the dataframe
```

```python
df.head(10)
```

```python
#check for null values
df.isna()
```

```python
#Remove null values
df = df.dropna()
```

```python
df.plot(kind='scatter', x='Unit price', y='gross income', color='blue')
plt.show()
```

```python
df.plot(kind='scatter', x='Unit price', y='Rating', color='red')
plt.show()
```

```python
sales_by_gender = df.groupby('Gender').count()['Invoice ID']
sales_by_gender
```

```python
sales_by_gender.plot(kind='bar', xlabel='Gender', ylabel='Number of Sales', title='Analysis of sales by gender')
```

```python
sales_by_city = df.groupby('City').count()['Invoice ID']
sales_by_city
```

```python
sales_by_city.plot(kind='bar', xlabel='City', ylabel='Number of Sales', title='Analysis of sales by city')
```

```python
sales_by_city.plot(kind='pie', ylabel='', title='Analysis of sales by city')
```

```python
sales_by_city.plot(kind='line', ylabel='', title='Analysis of sales by city')
```

```python
sales_by_city_total = df.groupby('City').count()[['Total', 'gross income']]
sales_by_city_total
```
